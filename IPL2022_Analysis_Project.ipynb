```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("/content/IPL.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
match_wins = df["match_winner"].value_counts()
```

```python
sns.barplot(y = match_wins.index, x = match_wins.values, palette = "viridis")
plt.title("No. of wins by each team")
```

2. **Toss Dicision Trends**

```python
sns.countplot(x = df["toss_decision"], palette = "rainbow")
plt.title("Toss Dicision Tren ds")
```

3. **Toss Winner vs Match Winner**

```python
(((df['toss_winner'] == df["match_winner"]).sum() / df.shape[0]) * 100).round(2)
```

```python
#OR
df2 = df[df["toss_winner"] == df["match_winner"]]
count = df2["match_id"].count()
percentage = (count * 100) / df.shape[0]
percentage.round(2)
```

***Key Player Performances***

1. **Most "Player of the Match" awards**

```python
df.head()
```

```python
count = df["player_of_the_match"].value_counts().head(10)
count
```

```python
sns.barplot(y = count.index, x = count.values, palette = "mako")
plt.title("Top 10 player of the match")
```

2. **Top scorers**

```python
high = df.groupby("top_scorer")["highscore"].sum().sort_values(ascending = False).head(5)
high
```

```python
sns.barplot(x = high.values, y = high.index, palette = "rainbow")
```

3. **10 Best bowling figures**

```python
df.head()
```

```python
df["highest_wickets"] = df["best_bowling_figure"].apply(lambda x : x.split("--")[0])
df["highest_wickets"] = df["highest_wickets"].astype(int)
top_bowlers = df.groupby("best_bowling")["highest_wickets"].sum().sort_values(ascending = False).head(10)
top_bowlers
```

```python
sns.barplot(x = top_bowlers.values, y = top_bowlers.index, palette = "rainbow")
```
